> Write a `Dict` class, which behaves the same as `dict`, but can be accessed through attributes, as follows:

```python
class Dict(dict):
    
    def __init__(self, **kw):
        super(Dict, self).__init__(**kw)
        
    def __getattr__(self, key):
        try:
            return self[key]
        except KeyError:
            raise AttributeError(r"'Dict' object has no attribute '%s' " % key)
        
    def __setattr__(self, key, value):
        self[key] = value
```

```python
d = Dict(a = 1, b = 2)
```

```python
d['a']
```

```python
d.a
```
